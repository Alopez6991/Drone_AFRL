```python
import importlib.util
import sys
import numpy as np
import matplotlib.pyplot as plt
# Define the file path
file_path = '/home/austin/Nonlinear_and_Data_Driven_Estimation/Drone/MPC_drone_03.py'

# Define the module name
module_name = 'MPC_drone_03'

# Load the module
spec = importlib.util.spec_from_file_location(module_name, file_path)
MPCD = importlib.util.module_from_spec(spec)
sys.modules[module_name] = MPCD
spec.loader.exec_module(MPCD)
```

```python
T_length=2000
dt=.01
# make vx sin wave
t=np.linspace(0,T_length*dt,T_length)
vx=np.full (T_length,1)


mpc=MPCD.Mpc_Smd(vx=vx,dt=dt, n_horizon=10, r_weight=0.0)
```

```python
X,U=mpc.run_mpc()
```

```python
# X[:,0]

plt.plot(t,X[:,0],label='sim vx')
plt.plot(t,vx,label='vx target',linestyle='--',alpha=.5)
plt.legend()
plt.xlim([0,20])
plt.ylim([.95,1.05])
plt.xlabel('time')
plt.ylabel('vx')
```

```python
plt.plot(t,X[:,0])
# get teh derivative of the position to get the velocity
X_dot=np.diff(X[:,1])/dt
plt.plot(t[1:],X_dot,linestyle='dashed')
```
